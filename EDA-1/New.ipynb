## Check for the Missing values

```python
df.isnull().sum()
```

### Extracting a list of columns having null Values

```python
[(x,x**2) for x in range(1,6)]
```

```python
[(features,df[features].isnull().sum()) for features in df.columns if df[features].isnull().sum() > 0]
```

### Code Explanation
- Here we use features as a variable to iterate over all the columns(features) in the dataset
- df[features] takes the column one by one from the dataset
- df[features].isnull().sum() gives the number of missing values in that specific column
- features is returned that is the list of columns in the dataset which have null values
<br/>
- This is known as Python List Comprehension, About which you can read more [here](https://platypusperry.medium.com/list-comprehension-in-python-96e2884c564)

### There is other file here named Country-Code.xlsx
It is a file that contains the country code for each country which maps the codes in the data-set to the country names.

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(3)
```

```python
final_df.dtypes #Check Data Types
```

Check number of entries in a column

```python
final_df.Country.value_counts()
```

## Storing the values to plot pie chart

```python
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
# Top 3 countries with most number of orders
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

#### Observation: Maximum number of orders are from India then comes United States and after it United Kingdom

```python
df_rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
df_rating.head(3)
```

Currently the whole row will be considered as index, so we need to reset the index to make a new index column.
Also rename the column showing counts to 'Rating Count' for better understanding.

```python
df_rating = df_rating.reset_index().rename(columns={0:'Rating Count'})
df_rating
```

```python
sns.barplot(x='Aggregate rating', y ='Rating Count',data=df_rating, palette=['white','red','orange','yellow','green','green'],hue='Rating color')
```

## Some General Queries

Getting Number of Users from Country who haven't rated yet

```python
final_df[final_df['Rating color']=='White'].groupby('Country').size().sort_values(ascending=False).head(10).reset_index().rename(columns={0:'Count'})
```

Currency name of each Country

```python
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index().rename(columns={0:'count'})
```

Which Country has online delivery available?

```python
final_df[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index().rename(columns={0:'count'})
```

Online Delivery is Available only in India and UAE nut also there are regions in both nations where online delivery is not available as well.
